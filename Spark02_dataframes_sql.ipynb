```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', inferSchema = True, header = True)
```

## sql way to show the results

```python
df.filter("Close < 500").show() #combine filter with sql syntax
```

```python
df.filter("Close < 500").select('Open').show()
```

```python
df.filter("Close < 500").select(['Open', 'Close']).show()
```

## Python, Pyspark way to show the results

```python
df.filter(df['Close'] < 500).show()
```

```python
df.filter(df['Close'] < 500).select('Volume').show()
```

```python
# the wrong way
# 1. df.filter(df['Close']< 200 and df['Open'] > 200).show()
# 2. df.filter(df['Close']< 200 & df['Open'] > 200).show()
```

```python
# the correct way: don't forget the paranthesis
df.filter((df['Close']< 200) & (df['Open'] > 200)).show()
```

```python
df.filter((df['Close']< 200) & ~(df['Open'] > 200)).show()
```

```python
df.filter(df['Low'] == 197.16).show()
```

```python
# to save it as a result
result = df.filter(df['Low'] == 197.16).collect()
```
