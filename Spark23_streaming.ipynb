# Pyspark: Streaming

```python
import findspark
findspark.init('/home/srabin/spark-3.5.1-bin-hadoop3')
import pyspark
from pyspark import SparkContext
```

```python
from pyspark.streaming import StreamingContext
```

```python
sc = SparkContext('local[2]', 'NetworkWordCount') # 2 local threads
```

```python
ssc = StreamingContext(sc, 1) # interval is 1 second
```

```python
lines = ssc.socketTextStream('localhost', 9999)
```

```python
words = lines.flatMap(lambda line: line.split(' ')) # create list of words
```

```python
pairs = words.map(lambda word: (word, 1))
```

```python
word_counts = pairs.reduceByKey(lambda num1, num2: num1 + num2)
```

```python
word_counts.pprint()
```

```python
ssc.start()
```
